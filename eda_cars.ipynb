# **Dataset Columns:**
- `Company`:The brand or manufacturer of the car (e.g., Ferrari, Ford, Audi).

- `Engine`:The engine type or layout (e.g., V8, V12, 1.2L Petrol).

- `Total Speed`:The top speed of the car (in km/h).

- `Price_Market`:price of the car (in US dollars).
- `Fuel Type`:Type of fuel the car uses (e.g., Petrol, Plug-in Hybrid).
- `Seats`:Total number of seats in the car.
- `Torque`:Rotational force generated by the engine (in Nm).

```python

# clean car_prices column (handles $, commas, different dashes, ranges) for accuracy of data
import re
def clean_value(cars_prices):
    if pd.isna(cars_prices):
        return pd.NA
    s = str(cars_prices).strip()
    # normalize dashes like â€“ â€” âˆ’ to simple hyphen
    s = re.sub(r'[â€“â€”âˆ’]', '-', s)
    # remove everything except digits, dot and hyphen
    s = re.sub(r'[^\d\.\-]', '', s)
    if not s:
        return pd.NA
    try:
        if '-' in s:
            parts = [p for p in s.split('-') if p]
            nums = [float(p) for p in parts]
            avg = sum(nums) / len(nums)
            return int(round(avg))
        else:
            return int(round(float(s)))
    except:
        return pd.NA

cols_to_fix = ['cars_prices']  #remember your columns name
for col in cols_to_fix:
    df[col] = df[col].apply(clean_value)
    # convert to pandas nullable integer dtype so .0 never show and NaNs stay as <NA>
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')
```

### **10- Acceleration Time (0â€“100)km/h vs Torque Category**

```python
# Convert performance column to numeric (remove non-numeric chars if any)
df['performance(0_-_100_)km/h'] = pd.to_numeric(
    df['performance(0_-_100_)km/h'].astype(str).str.replace(r'[^0-9.]', '', regex=True),errors='coerce')
# Torque bins
df['Torque_Category'] = pd.cut(df['torque'],
    bins=[0, 150, 250, 350, 500, df['torque'].max()],
    labels=['<150 Nm', '150â€“250 Nm', '250â€“350 Nm', '350â€“500 Nm', '500+ Nm'])
# Plot
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
#plotting violin plot
sns.violinplot(data=df,x='Torque_Category',y='performance(0_-_100_)km/h',palette='Set2',inner='quartile')
# Title & labels
plt.title("0â€“100 km/h Acceleration Time vs Torque Category", fontsize=16, fontweight='bold')
plt.xlabel("Torque Category", fontsize=12)
plt.ylabel("0â€“100 km/h Time (seconds)", fontsize=12)
# Set clean y-axis ticks
min_val = df['performance(0_-_100_)km/h'].min()
max_val = df['performance(0_-_100_)km/h'].max()
plt.yticks(np.arange(int(np.floor(min_val)), int(np.ceil(max_val))+1, 1))
#adjusting layout
plt.tight_layout()
plt.show()
```

### **Observations**
---
1: **Higher Torque faster Acceleration:** You may see that the 500+ Nm category has much lower acceleration times compared to <150 Nm indicating stronger torque improves performance.\
2: **Wider Spread in Mid-Torque Cars:** Categories like 250-50 Nm or 350-500 Nm may have a broader spread of acceleration times, showing variety in car weights, drivetrains, and tuning.\
3: **Entry-Level Cars are Slower:** <150 Nm and 150-250 Nm bins likely have higher median acceleration times, typical of economy-focused models.\
4: **Luxury & Performance Clustering:** 500+ Nm torque cars may have a tight cluster around low acceleration times (sports/luxury models).
